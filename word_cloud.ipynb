```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
```

```python
fake = pd.read_csv("data/Fake.csv", usecols= ["text"])
true = pd.read_csv("data/True.csv", usecols= ["text"])
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white',\
     stopwords = set(STOPWORDS), max_words = 200, max_font_size = 100, width = 800,\
          height = 400).generate(str(fake.text))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()

word_cloud2 = WordCloud(collocations = False, background_color = 'white',\
     stopwords = set(STOPWORDS), max_words = 200, max_font_size = 100, width = 800,\
          height = 400).generate(str(true.text))
plt.imshow(word_cloud2, interpolation='bilinear')
plt.axis("off")
plt.show()

```
